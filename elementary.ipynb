```python
import matplotlib.pyplot as plt
import numpy as np
import random

%matplotlib inline
```

In elementary cellular automata, the state of a cell is determined based off of the state of the three cells above it. The *rule* determines how the states of the three cells above correspond to the state of the cell below.

```python
def rule(m):
    a = bin(m)[2:]
    n = ('0' * (8 - len(a))) + a
    return {  '111': n[0]
            , '110': n[1]
            , '101': n[2]
            , '100': n[3]
            , '011': n[4]
            , '010': n[5]
            , '001': n[6]
            , '000': n[7] }
```

```python
def windows(parent):
    return [parent[i:i+3] for i in range(len(parent)-2)]
```

```python
def step(parent, rule):
    return '0'+''.join(rule[window] for window in windows(parent))+'0'
```

```python
def generate(init, rule, steps):
    out = ''
    curr = init
    for _ in range(steps):
        out = ''.join([out, curr, '\n'])
        curr = step(curr, rule)
    return out
```

```python
def visualize(automaton, cmap='binary'):
    tmp = automaton.strip().split('\n')
    data = np.array([[{'1':255, '0':0}[y] for y in x] for x in tmp])
    plt.imshow(data, cmap=cmap)
```

```python
def ecl(root, m, x, cmap='binary'):
    visualize(generate(root, rule(m), x), cmap=cmap)
```

```python
x = 2**8
root = '0'*x + '1' + '0'*x
ecl(root, 30, x)
```
